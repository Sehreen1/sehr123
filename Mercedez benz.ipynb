# APPLY LABEL ENCODER

```python
from sklearn.preprocessing import LabelEncoder
```

```python
x_train=train_obj.apply(LabelEncoder().fit_transform)
x_train
```

```python
x_test=test_obj.apply(LabelEncoder().fit_transform)
x_test
```

# DIMENSIONALITY REDUCTION

```python
y_train=df_train['y']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled_df=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled_df=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=0.90)
x_train_pca=pca.fit_transform(x_train_scaled_df)
x_test_pca=pca.transform(x_test_scaled_df)
```

# XG BOOST

```python
from sklearn import svm
from sklearn.metrics import r2_score, mean_squared_error
import math
import xgboost as xgb
from xgboost import XGBRegressor
from math import sqrt

```

```python
xg_boost=XGBRegressor()
xg_boost_pca=xg_boost.fit(x_train_pca,y_train)
train_pred=xg_boost.predict(x_train_pca)
test_pred=xg_boost.predict(x_test_pca)
```

```python
r2=r2_score(y_train,train_pred)
r2
```
