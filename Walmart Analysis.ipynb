CV:A coefficient of variation, often abbreviated as CV, is a way to measure how spread out values are in a dataset relative to the mean. It is calculated as:

CV = σ / μ

where:

σ: The standard deviation of dataset
μ: The mean of dataset
In plain English, the coefficient of variation is simply the ratio between the standard deviation and the mean.

**Which store/s has good quarterly growth rate in Q3’2012**

```python
df['Quarter']=df['Date'].dt.to_period('Q')
df.head()
```

```python
df2012=df[df['year']==2012]
dfq2=df2012[df2012['Quarter']=='2012Q2']
dfq3=df2012[df2012['Quarter']=='2012Q3']
dfq2
```

```python
Q3=pd.DataFrame(dfq3.groupby('Store')['Weekly_Sales'].sum())
Q3
```

```python
q3_growth=Q2.merge(Q3,how='inner',on='Store')
q3_growth.rename(columns={'Weekly_Sales_x': 'Q2_Weekly_Sales','Weekly_Sales_y':'Q3_Weekly_Sales'},inplace=True)
```

```python
q3_growth
```

```python
q3_growth['Growth Rate']=(q3_growth['Q3_Weekly_Sales'] - q3_growth['Q2_Weekly_Sales'])/q3_growth['Q2_Weekly_Sales']
q3_growth['Growth Rate']=round(q3_growth['Growth Rate'],2)
q3_growth.sort_values('Growth Rate',ascending=False).head(1)
```

# MODEL BUILDING

finding corelation between dependent variables and independent variable
weekly sales is our independent variable and consider dependent variables as  CPI, unemployment, and fuel price


```python
df.columns
```

```python
df1=df[['Fuel_Price', 'CPI', 'Unemployment','Weekly_Sales']]
```

```python
df1.corr()['Weekly_Sales']
```

```python
import statsmodels.formula.api as sm
model=sm.ols('Weekly_Sales~Fuel_Price+ CPI+ Unemployment',data=df1).fit()
```

```python
print(model.summary())
```

Observations:
p value of Fuel_Price is greater;therefore we will try to build model without considering fuel price.


```python
model2=sm.ols('Weekly_Sales~ CPI+ Unemployment',data=df1[['CPI','Unemployment','Weekly_Sales']]).fit()
```

```python
print(model2.summary())
```

```python
model2.pvalues
```

p values for CPI and unemployment have p values less tha 0.05;
This indicates  and cpi and unemployment are both significant parameters

# PREDICTION

```python
df3=df[[ 'CPI', 'Unemployment','Weekly_Sales']]
pred=model2.predict(df3)
```

**accuracy:**
    

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(df3['Weekly_Sales'],pred)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```
